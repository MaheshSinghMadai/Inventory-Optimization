```python
# Warehous Inventory Optimization : Avoiding overstocking problem by demannd forecasting  and 
# setting inventory levels based on predicted demand and safety stock
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from xgboost import XGBRegressor
import os
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load and preprocess the UCI Online Retail Dataset

df = pd.read_csv('online_retail.csv',encoding='latin-1')
df = df.dropna(subset=['StockCode', 'Quantity', 'InvoiceDate'])
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df = df[df['Quantity'] > 0]  # Exclude returns



# Aggregate daily sales per StockCode

daily_sales = df.groupby(['StockCode', pd.Grouper(key='InvoiceDate', freq='D')])['Quantity'].sum().reset_index()
print(f"Dataset shape: {daily_sales.shape}")
print(daily_sales.head())
```

```python
# Ensure necessary directories exist

forecasts_dir = 'forecasts'
metrics_dir = 'metrics'
os.makedirs(forecasts_dir, exist_ok=True)
os.makedirs(metrics_dir, exist_ok=True)

# Prepare data and select sample StockCodes
stockcodes = daily_sales['StockCode'].unique()
forecast_horizon = 7  # 7-day forecast

# Select top 10 StockCodes by frequency
stockcode_counts = daily_sales['StockCode'].value_counts()
sample_stockcodes = stockcode_counts.head(10).index.tolist()
print(f"Processing {len(sample_stockcodes)} StockCodes for visualization and forecasting.")

# Function to train models and generate forecasts
def train_and_forecast(stockcode, stock_data):
    if len(stock_data) < 2:
        print(f"Skipping {stockcode}: Insufficient data.")
        return None, None, None
    
    # ARIMA Model
    try:
        model_arima = ARIMA(stock_data, order=(1, 1, 1))
        results_arima = model_arima.fit()
        forecast_arima = results_arima.forecast(steps=forecast_horizon)
        forecast_dates = pd.date_range(start=stock_data.index[-1] + timedelta(days=1), 
                                       periods=forecast_horizon, freq='D')
        arima_forecast = pd.DataFrame({'Date': forecast_dates, 'predicted_sales': forecast_arima})
    except Exception as e:
        print(f"ARIMA failed for {stockcode}: {e}")
        return None, None, stock_data
    
    # XGBoost Model
    try:
        stock_data_df = stock_data.reset_index()
        stock_data_df['day'] = stock_data_df['InvoiceDate'].dt.day
        stock_data_df['month'] = stock_data_df['InvoiceDate'].dt.month
        stock_data_df['year'] = stock_data_df['InvoiceDate'].dt.year
        stock_data_df['lag1'] = stock_data_df['Quantity'].shift(1)
        stock_data_df = stock_data_df.dropna()

        X = stock_data_df[['day', 'month', 'year', 'lag1']]
        y = stock_data_df['Quantity']
        model_xgb = XGBRegressor(n_estimators=100, random_state=42)
        model_xgb.fit(X, y)

        last_date = stock_data_df['InvoiceDate'].iloc[-1]
        future_dates = [last_date + timedelta(days=i) for i in range(1, forecast_horizon + 1)]
        future_features = pd.DataFrame({
            'day': [d.day for d in future_dates],
            'month': [d.month for d in future_dates],
            'year': [d.year for d in future_dates],
            'lag1': [stock_data_df['Quantity'].iloc[-1]] * forecast_horizon
        })
        forecast_xgb = model_xgb.predict(future_features)
        xgb_forecast = pd.DataFrame({'Date': future_dates, 'predicted_sales': forecast_xgb})
    except Exception as e:
        print(f"XGBoost failed for {stockcode}: {e}")
        return arima_forecast, None, stock_data
    
    return arima_forecast, xgb_forecast, stock_data
```

```python
# Visualize demand patterns

plt.figure(figsize=(15, 10))
for i, stockcode in enumerate(sample_stockcodes, 1):
    plt.subplot(4, 3, i)
    stock_data = daily_sales[daily_sales['StockCode'] == stockcode].set_index('InvoiceDate')['Quantity']
    plt.plot(stock_data.index, stock_data, label=f'StockCode {stockcode}')
    plt.title(f'Demand Pattern for StockCode {stockcode}')
    plt.xlabel('Date')
    plt.ylabel('Quantity')
    plt.legend()
    plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Train models and collect forecasts

arima_forecasts = {}
xgb_forecasts = {}
historical_data = {}

for stockcode in sample_stockcodes:
    stock_data = daily_sales[daily_sales['StockCode'] == stockcode].set_index('InvoiceDate')['Quantity']
    arima_fc, xgb_fc, hist_data = train_and_forecast(stockcode, stock_data)
    if arima_fc is not None:
        arima_forecasts[stockcode] = arima_fc
    if xgb_fc is not None:
        xgb_forecasts[stockcode] = xgb_fc
    if hist_data is not None:
        historical_data[stockcode] = hist_data

print(f"Forecasts generated for {len(sample_stockcodes)} StockCodes.")

# Task 2: Evaluate Models
rmse_arima = []
rmse_xgb = []
mape_arima = []
mape_xgb = []

for stockcode in sample_stockcodes:
    hist_data = historical_data.get(stockcode)
    if (hist_data is not None and len(hist_data) > forecast_horizon and 
        stockcode in arima_forecasts and stockcode in xgb_forecasts):
        actual = hist_data[-forecast_horizon:].values
        arima_pred = arima_forecasts[stockcode]['predicted_sales'].values
        xgb_pred = xgb_forecasts[stockcode]['predicted_sales'].values
        rmse_arima.append(np.sqrt(mean_squared_error(actual, arima_pred)))
        rmse_xgb.append(mean_absolute_percentage_error(actual, xgb_pred))
        mape_arima.append(np.sqrt(mean_squared_error(actual, arima_pred)))
        mape_xgb.append(mean_absolute_percentage_error(actual, xgb_pred))

# Average metrics
avg_rmse_arima = np.mean(rmse_arima) if rmse_arima else 0
avg_rmse_xgb = np.mean(rmse_xgb) if rmse_xgb else 0
avg_mape_arima = np.mean(mape_arima) if mape_arima else 0
avg_mape_xgb = np.mean(mape_xgb) if mape_xgb else 0

print(f"Average RMSE - ARIMA: {avg_rmse_arima:.2f}, XGBoost: {avg_rmse_xgb:.2f}")
print(f"Average MAPE - ARIMA: {avg_mape_arima:.2f}%, XGBoost: {avg_mape_xgb:.2f}%")
```

```python
# Visualization of metrics

models = ['ARIMA', 'XGBoost']
x = np.arange(len(models))
width = 0.35

plt.figure(figsize=(12, 5))

# RMSE Chart
plt.subplot(1, 2, 1)
plt.bar(x - width/2, [avg_rmse_arima, avg_rmse_xgb], width, color=['skyblue', 'lightgreen'])
plt.xlabel('Models')
plt.ylabel('RMSE')
plt.title('Root Mean Squared Error Comparison')
plt.xticks(x, models)
plt.grid(True)

# MAPE Chart
plt.subplot(1, 2, 2)
plt.bar(x - width/2, [avg_mape_arima, avg_mape_xgb], width, color=['skyblue', 'lightgreen'])
plt.xlabel('Models')
plt.ylabel('MAPE (%)')
plt.title('Mean Absolute Percentage Error Comparison')
plt.xticks(x, models)
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Export predictions for top 10 most frequent StockCode

for stockcode in sample_stockcodes:
    if stockcode in arima_forecasts:
        arima_forecasts[stockcode].to_csv(
            os.path.join(forecasts_dir, f'arima_forecast_{stockcode}.csv'), index=False)
    if stockcode in xgb_forecasts:
        xgb_forecasts[stockcode].to_csv(
            os.path.join(forecasts_dir, f'xgb_forecast_{stockcode}.csv'), index=False)

print(f"Forecasts exported for StockCode {most_frequent_stockcode} to {forecasts_dir}")

# Task 4: Export performance metrics to CSV
metrics_data = {
    'Model': ['ARIMA', 'XGBoost'],
    'Average_RMSE': [avg_rmse_arima, avg_rmse_xgb],
    'Average_MAPE (%)': [avg_mape_arima, avg_mape_xgb]
}

try:
    metrics_df = pd.DataFrame(metrics_data)
    output_path = os.path.join(metrics_dir, 'performance_metrics.csv')
    metrics_df.to_csv(output_path, index=False)
    print(f"Performance metrics exported to {output_path}")
except Exception as e:
    print(f"Error exporting performance metrics: {e}")

if not os.access(metrics_dir, os.W_OK):
    print(f"Warning: No write permission for directory {metrics_dir}")
```
